```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
df=pd.read_csv('C:/Users/82102/Desktop/문화 관광 project/지역 코드.csv', encoding='utf-8')
```

```python
df.rename(columns={'시/도명':'시도명',
                  '시/군/구명':'시군구명',
                  '시/도코드':'시도코드',
                  '시/군/구코드':'시군구코드'}, inplace=True)
```

```python
df
```

```python
df['방문지역코드']=0
for i in range(len(df)):
    if len(str(df['시군구코드'][i])) == 2:
        df['방문지역코드'][i]=int(str(df['시도코드'][i]) + '0' + str(df['시군구코드'][i]))
    else:
        df['방문지역코드'][i]=int(str(df['시도코드'][i]) + str(df['시군구코드'][i]))
```

```python
df['지역명']=''
for i in range(len(df)):
    if len(df['시도명'][i]) == '세종특별자치시' :
        df['지역명'][i]=df['시도명'][i][:1]
    elif len(df['시도명'][i]) == 4:
        df['지역명'][i]=df['시도명'][i][::2]+' '+df['시군구명'][i]
    else:
        df['지역명'][i]=df['시도명'][i][:2]+' '+df['시군구명'][i]
```

```python
df=df[['방문지역코드','시도명','시군구명','지역명']]
```

```python
df.to_csv('지역코드정리.csv', index=False, encoding='UTF-8-sig')
```
